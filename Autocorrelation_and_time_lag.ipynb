```
import pandas as pd
import matplotlib.pyplot as plt
f=pd.read_csv("/content/corr.csv.csv")
```

**Convert the data type to variable float**

```
f['t0']=pd.to_numeric(f['t0'],downcast='float')
f['t0']
```

**plot the autocorrelation**

```
plt.acorr(f['t0'],maxlags=10)
```

**Create the time lag dataset and concatenate the result**

```
t_1=f['t0'].shift(+1).to_frame()
t_1
```

```
t_1.columns=['t-1']
t_1.columns
t_1
```

```
t_2=f['t0'].shift(+2).to_frame()
t_2.columns=['t-2']
t_2
```

```
result=pd.concat([f['t0'],t_1,t_2],axis=1)
result
```
